# Демонстрация библиотеки `whether`

Библиотека предназначена для загрузки, импорта/экспорта и анализа суточных метеорологических данных через API Open-Meteo.

---

## 1. Импорт и настройка

````python
import sys
from pathlib import Path

ROOT = Path().resolve().parents[0]
sys.path.append(str(ROOT / "src"))

from whether import Whether
````

````python
# Координаты Санкт-Петербурга
LATITUDE = 59.93
LONGITUDE = 30.33
````

---
## 2. Создание объекта Whether

При инициализации передаём координаты точки наблюдения и часовой пояс.

````python
weather = Whether(latitude=LATITUDE, longitude=LONGITUDE, timezone="Europe/Moscow")
````

---
## 3. Загрузка данных из API

Метод `download_whether_data` обращается к Open-Meteo Archive API и возвращает DataFrame с суточными метеопараметрами.

````python
df_jan = weather.download_whether_data(
    start_date="2025-01-01",
    end_date="2025-01-31",
)
df_jan.head()
````

````python
# Догрузим февраль — данные автоматически объединятся,
# дубликаты по дате будут удалены
df_feb = weather.download_whether_data(
    start_date="2025-02-01",
    end_date="2025-02-28",
)
print(f"Всего строк после объединения: {len(weather.whether_data)}")
````

````python
weather.whether_data
````

## 4. Экспорт данных

Библиотека поддерживает экспорт через `ExportWhetherFactory`, которая определяет формат по расширению файла.

Поддерживаемые форматы:

- **CSV** — при передаче пути с расширением `.csv`
  ```python
  weather.export_whether_data("data/weather_spb.csv")
  ```
- **Excel** — при расширении .xlsx или .xls
  ```python
  weather.export_whether_data("data/weather_spb.xlsx")
  ```
- **JSON** — при расширении .json (список объектов, orient="records")
  ```python
  weather.export_whether_data("data/weather_spb.json")
  ```

Формат выбирается автоматически по расширению файла.

````python
# Экспорт в CSV
weather.export_whether_data("weather_spb.csv")
````

````python
# Экспорт в Excel
weather.export_whether_data("weather_spb.xlsx")
````

````python
# Экспорт в JSON
weather.export_whether_data("weather_spb.json")
````

## 5. Импорт данных из файлов

Создадим новый объект и загрузим данные из ранее сохранённых файлов.

````python
weather_from_files = Whether(latitude=LATITUDE, longitude=LONGITUDE)

# Чтение из CSV
weather_from_files.read_whether_data("weather_spb.csv")
print(f"Строк после чтения CSV: {len(weather_from_files.whether_data)}")
````

````python
# Дочитаем тот же период из Excel — дубликаты не появятся
weather_from_files.read_whether_data("weather_spb.xlsx")
print(f"Строк после добавления Excel: {len(weather_from_files.whether_data)}")
````

````python
# Проверим, что дубликатов нет
has_dups = weather_from_files.whether_data.duplicated(subset=["date"]).any()
print(f"Есть дубликаты по дате: {has_dups}")
````

## 7. Пользовательские исключения

Библиотека выбрасывает ошибки при некорректных входных данных.

- `InvalidCoordinateError` - Бросается, если широта или долгота выходят за допустимые пределы системы координат WGS84.

    Когда может случиться:
    - Опечатка в координатах (latitude=123.0, longitude=250.0).
    - Перепутаны местами широта/долгота.

- `InvalidTimezoneError` - Бросается, если строка timezone не является валидным именем таймзоны из базы IANA.

    Когда может случиться:
    - Пользователь передал `"Moscow"` вместо `"Europe/Moscow"`.
    - Опечатка в названии временной зоны.

- `InvalidDateFormatError` - Сигнализирует, что строковая дата либо задана в неверном формате, либо не существует как календарная дата.

    Невозможные даты: `"2025-02-31"`, `"2025-13-01"`, `"2025-1-01"`, `"25-01-01"`, `"2025-01-1"`.
